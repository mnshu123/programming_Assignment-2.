```python
#Q1.   Write a Python program to convert kilometers to miles?

km = float(input())
miles = km* 0.621371
print("Km into Miles:",miles)
```

```python
#Q2.   Write a Python program to convert Celsius to Fahrenheit?

Celsius = float(input())
Fahrenheit = (Celsius*9/5)+32

print("Celsius into Fahrenheit:",Fahrenheit)
```

```python
#Q3.  Write a Python program to display calendar?

import calendar
year = int(input("year"))
month = int(input("Month"))
print(calendar.month(year,month))
```

```python
#Q4.  Write a Python program to solve quadratic equation?
import cmath

num1 = float(input("num1:"))
num2 = float(input("num2:"))
num3 = float(input("num3:"))

formula = num2**2-4*num1*num3
Positive = (-num2+ cmath.sqrt(formula))/(2*num1)
Negative = (-num2- cmath.sqrt(formula))/(2*num1)

print("Positive:",Positive)
print("Nagetive:",Negative)
```
